```python
import numpy as np
import matplotlib.pyplot as plt
import math

from scipy.special import comb
```

# 0. Sanity check surrounding maximum entropy state

```python
N = 50 # make sure this number is even

N_R = np.arange(0, N+1)
N_L = N - N_R

g_L = comb(N, N_L)
g_R = comb(N, N_R)

S_tot = np.log(g_L*g_R)

S_tot_approx = 2*N*np.log(N) - 2*N_R*np.log(N_R) - 2*N_L*np.log(N_L)

plt.plot(N_R, S_tot)
plt.plot(N_R, S_tot_approx, c="r")
plt.title("Total entropy")
plt.xlabel("$N_R$")
plt.show()

dSdN_R = np.zeros(N-1)
for i in range(N-1):
    dSdN_R[i] = S_tot[i+1]-S_tot[i]

plt.plot(N_R[:-2], dSdN_R)
plt.axvline(N/2)
#plt.xlim(0.5, 1)
#plt.ylim(top=0)
plt.ylim
plt.show()

```
